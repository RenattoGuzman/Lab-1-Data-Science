```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Definir explícitamente las columnas categóricas
categorical_columns = [
    'Smokes', 'STDs.molluscum.contagiosum', 'STDs.AIDS', 'STDs.HIV', 'STDs.Hepatitis.B',
    'STDs.HPV', 'Dx.Cancer', 'Dx.CIN', 'Dx.HPV', 'Dx', 'Hinselmann', 'Schiller', 'Citology',
    'Biopsy', 'Hormonal.Contraceptives', 'IUD', 'STDs', 'STDs.condylomatosis',
    'STDs.cervical.condylomatosis', 'STDs.vaginal.condylomatosis',
    'STDs.vulvo.perineal.condylomatosis', 'STDs.syphilis', 'STDs.pelvic.inflammatory.disease',
    'STDs.genital.herpes'
]

# Convertir estas columnas a categóricas (incluso si tienen valores numéricos)
for column in categorical_columns:
    if column in data.columns:
        data[column] = data[column].astype('category')

# Verificar las variables categóricas
categorical_vars = data.select_dtypes(include=['category'])

print("\nVariables categóricas identificadas:")
print(categorical_vars.columns)

# Crear tablas de frecuencia y proporción para cada variable categórica
for column in categorical_vars.columns:
    freq_table = data[column].value_counts(dropna=False)
    prop_table = data[column].value_counts(normalize=True, dropna=False)

    print(f"\nFrecuencia y proporción de {column}:")
    display(pd.DataFrame({'Frecuencia': freq_table, 'Proporción': prop_table}))

    # Gráfico de barras para cada variable categórica
    plt.figure(figsize=(8, 6))
    sns.countplot(x=column, data=data, order=freq_table.index)
    plt.title(f'Distribución de {column} (Barra)')
    plt.xlabel(column)
    plt.ylabel('Frecuencia')
    plt.show()

    # Gráfico de pastel para cada variable categórica
    plt.figure(figsize=(8, 6))
    freq_table.plot.pie(autopct='%1.1f%%', startangle=90)
    plt.title(f'Distribución de {column} (Pastel)')
    plt.ylabel('')
    plt.show()

    # Gráfico de conteo apilado para cada variable categórica (si hay más de una categoría)
    if len(freq_table) > 1:
        plt.figure(figsize=(8, 6))
        freq_table.plot(kind='bar', stacked=True)
        plt.title(f'Distribución apilada de {column}')
        plt.xlabel(column)
        plt.ylabel('Frecuencia')
        plt.show()

    # Gráfico de conteo agrupado para cada variable categórica (si hay más de una categoría)
    if len(freq_table) > 1:
        plt.figure(figsize=(8, 6))
        sns.countplot(x=column, data=data, order=freq_table.index, hue=column)
        plt.title(f'Distribución agrupada de {column}')
        plt.xlabel(column)
        plt.ylabel('Frecuencia')
        plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.decomposition import PCA
from sklearn.impute import SimpleImputer

# Cargar el dataset
data = pd.read_csv("risk_factors_cervical_cancer.csv")

# Reemplazar '?' con NaN y convertir a numérico
data.replace('?', np.nan, inplace=True)
data = data.apply(pd.to_numeric, errors='ignore')

# Definir explícitamente las columnas categóricas
categorical_columns = [
    'Smokes', 'STDs.molluscum.contagiosum', 'STDs.AIDS', 'STDs.HIV', 'STDs.Hepatitis.B',
    'STDs.HPV', 'Dx.Cancer', 'Dx.CIN', 'Dx.HPV', 'Dx', 'Hinselmann', 'Schiller', 'Citology',
    'Biopsy', 'Hormonal.Contraceptives', 'IUD', 'STDs', 'STDs.condylomatosis',
    'STDs.cervical.condylomatosis', 'STDs.vaginal.condylomatosis',
    'STDs.vulvo.perineal.condylomatosis', 'STDs.syphilis', 'STDs.pelvic.inflammatory.disease',
    'STDs.genital.herpes'
]

# Codificación de Etiquetas
label_encoder = LabelEncoder()
for column in categorical_columns:
    if column in data.columns:
        data[column] = label_encoder.fit_transform(data[column].astype(str))

# Imputar valores faltantes
imputer = SimpleImputer(strategy='mean')  # Puedes cambiar la estrategia según sea necesario
data_imputed = pd.DataFrame(imputer.fit_transform(data), columns=data.columns)

# Estandarización de todas las variables
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_imputed)

# Aplicar PCA
pca = PCA(n_components=2)  # Ajusta el número de componentes principales según sea necesario
principal_components = pca.fit_transform(data_scaled)

# Convertir los componentes principales a un DataFrame
pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])

# Graficar los componentes principales
plt.figure(figsize=(8, 6))
sns.scatterplot(x='PC1', y='PC2', data=pca_df)
plt.title('Componentes Principales')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()

# Mostrar la varianza explicada por cada componente
print("Varianza explicada por componente:", pca.explained_variance_ratio_)
```
